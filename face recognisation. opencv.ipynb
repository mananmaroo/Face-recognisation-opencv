```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=detector.detectMultiScale(img)
```

```python
for face in faces:
    x,y,w,h = face
    modified_img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)
```
